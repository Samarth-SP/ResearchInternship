```python
import pandas as pd
data = pd.read_json('arxiv-metadata-oai-snapshot.json', lines=True)
```

```python
data.head()
```

```python
print(data.columns)
```

```python
data = data.drop('report-no', axis=1)
```

```python
data = data.drop('submitter', axis=1)
data = data.drop('authors', axis=1)
data = data.drop('comments', axis=1)
data = data.drop('journal-ref', axis=1)
data = data.drop('versions', axis=1)
data = data.drop('id', axis=1)
data = data.drop('license', axis=1)
data = data.drop('update_date', axis=1)
data = data.drop('authors_parsed', axis=1)
```

```python
data.head()
```

```python
import missingno as msno
import matplotlib.pyplot as plt
msno.matrix(data);
plt.show()
msno.bar(data);
plt.show()
```

```python
data = data.dropna(how='any',axis=0) 
```

```python
data.info()
```

```python
data.head()
```

```python
data.reset_index(drop=True)
```

```python
print(data['categories'].value_counts())
```

```python
pd.DataFrame.to_json(data, path_or_buf='ArXiv.json', orient=None, date_format=None, double_precision=10, force_ascii=True, lines=False)
```
