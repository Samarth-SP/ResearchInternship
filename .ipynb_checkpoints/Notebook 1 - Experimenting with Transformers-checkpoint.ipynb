```python
import pandas as pd
import numpy as np
from transformers import pipeline
```

```python
hammy=open('hamlet.txt')
textgen = pipeline("sentiment-analysis")
results = []
for line in hammy:
    results.extend(textgen(line))
```

```python
for result in results:
    if result['label'] == "NEGATIVE":
        print((-1)*result['score'])
    else:
        print(result['score'])
```
